```python
def age_category(Age):
  if Age < 30:
   return 'young'
  elif 30 <= Age <= 50:
   return 'middle age'
  elif Age > 50:
   return 'old'
# Ñ‚ÐµÐ¿ÐµÑ€ÑŒ Ð¿Ñ€Ð¸ÑÐ¾ÐµÐ´Ð¸Ð½ÑÑŽ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ Ð¿Ñ€Ð¾Ð¿Ð¸ÑÐ½Ð½ÑƒÑŽ Ð² def Ðº Ð¸Ð·Ð½Ð°Ñ‡Ð°Ð»ÑŒÐ½Ð¾Ð¼Ñƒ ÑÑ‚Ð¾Ð»Ð±Ñ†Ñƒ Ð²Ð¾Ð·Ñ€Ð°ÑÑ‚Ð° Ð¸ Ð½Ð°Ð·Ñ‹Ð²Ð°ÑŽ Ð½Ð¾Ð²Ñ‹Ð¹ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† df['category_age']
#.apply('category_age') - ÑÑÑ‹Ð»Ð°ÑŽÑÑŒ Ð½Ð° ÐºÐ°Ñ‚ÐµÐ³Ð¾Ñ€Ð¸ÑŽ Ð²Ð¾Ð·Ñ€Ð°ÑÑ‚Ð°, Ðº ÑÑ‚Ð¾Ð»Ð±Ñ†Ñƒ age_category Ð¿Ñ€Ð¸Ð¼ÐµÐ½Ð¸ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ apply
df['category_age'] = df['Age'].apply(age_category)
display(df)
```

```python
from ast import In

def season(month):
  if month in [12,1,2]:
   return 'winter'
  elif month in [3,4,5]:
    return 'spring'
  elif month in [6,7,8]:
    return 'summer'
  elif month in [9,10,11]:
    return 'autumn'
df['season'] = df['mouth'].apply(season)
display(df)
```

```python
#rfm Ð°Ð½Ð°Ð»Ð¸Ð· Ð¿Ð¾ ÐºÐ°Ð¶Ð´Ð¾Ð¼Ñƒ ÐºÐ»Ð¸ÐµÐ½Ñ‚Ñƒ Ð¾Ñ‚Ð´ÐµÐ»ÑŒÐ½Ð¾.
import pandas as pd

# 1 Ñ€Ð°ÑÑÑ‡Ð¸Ñ‚Ð°Ñ‚ÑŒ ÐºÐ¾Ð»-Ð²Ð¾ Ð´Ð½ÐµÐ¹ Ð¾Ñ‚ Ð°Ð½Ð°Ð»Ð¸Ð·Ð° Ð´Ð¾ Ð¿Ð¾ÑÐ»ÐµÐ´Ð½ÐµÐ¹ Ð¿Ð¾ÐºÑƒÐ¿ÐºÐ¸.
df['order_recency'] = pd.to_datetime('2024-01-01') - pd.to_datetime(df['Date'].dt.date)

# 2 ÐÐµÐ¾Ð±Ñ…Ð¾Ð´Ð¸Ð¼Ð¾ ÑÐ³Ñ€ÑƒÐ¿Ð¿Ð¸Ñ€Ð¾Ð²Ð°Ñ‚ÑŒ Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð³Ð¾ Ð¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ñ‚ÐµÐ»Ñ Ð¸ Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð´Ð»Ñ RFM Ð°Ð½Ð°Ð»Ð¸Ð·Ð°.

rfm = df.groupby('Customer ID').agg(
    recency = ('order_recency', lambda x: x.min().days),
    frequeny = ('Transaction ID', 'nunique'),
    monetary_value = ('Total Amount', 'sum')
).reset_index()

# 3 Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»Ð¸Ñ‚ÑŒ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€ Ñ€Ð°Ð·Ð±Ð¸Ð²ÐºÐ¸ Ð¿Ð¾ Ð³Ñ€ÑƒÐ¿Ð¿Ð°Ð¼ Ð¿Ð¾ ÐºÐ°Ð¶Ð´Ð¾Ð¼Ñƒ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÑƒ.

# Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ RFM ÑÐµÐ³Ð¼ÐµÐ½Ñ‚Ð¾Ð²
rfm['r'] = pd.qcut(rfm['recency'], q=3, labels=[3, 2, 1])
rfm['f'] = pd.cut(rfm['frequeny'], [0, 2, 10, 1000], labels=[1, 2, 3])
rfm['m'] = pd.qcut(rfm['monetary_value'], q=3, labels=[1, 2, 3])

display(rfm)


# 4 Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ rfm Ð¸Ð´Ð½ÐµÐºÑÐ° ÑÐµÐ³Ð¼ÐµÐ½Ñ‚Ð°

rfm[['r','f','m']] = rfm[['r','f','m']].astype('str')
rfm['rfm_group'] = rfm['r'] + rfm['f'] + rfm['m']

rfm[['r','f','m']] = rfm[['r','f','m']].astype('int')
rfm['rfm_sum'] = rfm[['r','f','m']].sum(axis =1)

# 5 ÐÐµÐ¾Ð±Ñ…Ð¾Ð´Ð¸Ð¼Ð¾ ÑÐ³Ñ€ÑƒÐ¿Ð¿Ð¸Ñ€Ð¾Ð²Ð°Ñ‚ÑŒ Ð´Ð°Ð½Ð½Ñ‹Ðµ Ð¿Ð¾ ÑÐµÐ³Ð¼ÐµÐ½Ñ‚Ð°Ð¼ Ð¸ Ð¿Ð¾ÑÑ‡Ð¸Ñ‚Ð°Ñ‚ÑŒ Ð¸Ñ… Ñ€Ð°Ð·Ð¼ÐµÑ€

rfm_group = rfm.groupby('rfm_group').agg(
    {
        'Customer ID' : 'nunique',
        'rfm_sum' : 'mean'
    }
).reset_index()

display(rfm_group.sort_values(by = 'Customer ID'))
```
